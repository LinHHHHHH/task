```python
    import jaqs_fxdayu
    jaqs_fxdayu.patch_all()
    from jaqs.data import DataView
    from jaqs.data import RemoteDataService
    from jaqs_fxdayu.data.dataservice import LocalDataService
    import os
    import numpy as np
    import pandas as pd
    import warnings

    warnings.filterwarnings("ignore")

    dataview_folder = r'D:\股票因子数据\data'
    dv = DataView()
    ds = LocalDataService(fp=dataview_folder)


    factor_list = ['volume']
    check_factor = ','.join(factor_list)

    dv_props = {'start_date': start, 'end_date': end, 'symbol':','.join(stock_symbol),
             'fields': check_factor,
             'freq': 1,
             "prepare_fields": True}

    dv.init_from_config(dv_props, data_api=ds)
    dv.prepare_data()
```

```python
#1、MTM_2
dv.add_formula('MTM_1','close - Delay(close,12)',is_quarterly=False,add_data=True)
dv.add_formula('MTM_2','-Ewma(MTM_1,6)',is_quarterly=False,add_data=True)
```

```python
#2、MTR_2
dv.add_formula('MTR_1','Max(Max((high-low),Abs(Delay(close,1)-high)),Abs(Delay(close,1)-low))',is_quarterly=False,add_data=True)
dv.add_formula('MTR_2','-Ewma(MTR_1,14)',is_quarterly=False,add_data=True)
```

```python
#3、MAADTM_1
dv.add_formula('DTM_1','If(open<=Delay(open,1),0,Max((high-open),(open-Delay(open,1))))',is_quarterly=False,add_data=True)
dv.add_formula('DBM_1','If(open>=Delay(open,1),0,Max((open-low),(open-Delay(open,1))))',is_quarterly=False,add_data=True)
dv.add_formula('STM_1','Ts_Sum(DTM_1,23)',is_quarterly=False,add_data=True)
dv.add_formula('SBM_1','Ts_Sum(DBM_1,23)',is_quarterly=False,add_data=True)
dv.add_formula('ADTM_1','If(STM_1>SBM_1,(STM_1-SBM_1)/STM_1,If(STM_1==SBM_1,0,(STM_1-SBM_1)/SBM_1))',is_quarterly=False,add_data=True)
dv.add_formula('MAADTM_1','-Ewma(ADTM_1,8)',is_quarterly=False,add_data=True)
```

```python
#4、RSI6_1
dv.add_formula('DIF_1','close-Delay(close,1)',is_quarterly=False,add_data=True)
dv.add_formula('VU_1','If(DIF_1>=0,DIF_1,0)',is_quarterly=False,add_data=True)
dv.add_formula('VD_1','If(DIF_1<0,-DIF_1,0)',is_quarterly=False,add_data=True)
dv.add_formula('MAU1_1','Ewma(VU_1,10)',is_quarterly=False,add_data=True)
dv.add_formula('MAD1_1','Ewma(VD_1,10)',is_quarterly=False,add_data=True)
dv.add_formula('MAU2_1','Ewma(VU_1,6)',is_quarterly=False,add_data=True)
dv.add_formula('MAD2_1','Ewma(VD_1,6)',is_quarterly=False,add_data=True)
dv.add_formula('RSI10_1','Ewma(100*MAU1_1/(MAU1_1+MAD1_1),10)',is_quarterly=False,add_data=True)
dv.add_formula('RSI6_1','-(Ewma(100*MAU2_1/(MAU2_1+MAD2_1),6))',is_quarterly=False,add_data=True)
```

```python
#5、MAVPT_1
dv.add_formula('VPT_1','Ts_Sum(volume*(close-Delay(close,1))/Delay(close,1),51)',is_quarterly=False,add_data=True)
dv.add_formula('MAVPT_1','-Ts_Mean(VPT_1,6)',is_quarterly=False,add_data=True)
```

```python
#6、DIFMA_1
dv.add_formula('DIF_11','(Ts_Mean(close,10)-Ts_Mean(close,50))',is_quarterly=False,add_data=True)
dv.add_formula('DIFMA_1','-(Ts_Mean(DIF_11,10))',is_quarterly=False,add_data=True)
```

```python
#7、MADPO1
dv.add_formula('DPO1','(close-Delay(Ts_Mean(close,20),11))',is_quarterly=False,add_data=True)
dv.add_formula('MADPO1','-Ts_Mean(DPO1,6)',is_quarterly=False,add_data=True)
```

```python
#8、MAEMV1
dv.add_formula('VOLUME11','Ts_Mean(volume,14)/volume',is_quarterly=False,add_data=True)
dv.add_formula('MID11','100*(high+low-Delay(high+low,1))/(high+low)',is_quarterly=False,add_data=True)
dv.add_formula('EMV1','Ts_Mean(MID11*VOLUME11*(high-low)/Ts_Mean(high-low,14),14)',is_quarterly=False,add_data=True)
dv.add_formula('MAEMV1','-Ts_Mean(EMV1,9)',is_quarterly=False,add_data=True)
```

```python
#9、DDIF2
dv.add_formula('DIF2','Ewma(close,12)-Ewma(close,26)',is_quarterly=False,add_data=True)
dv.add_formula('MACD2','Ewma(DIF2,9)',is_quarterly=False,add_data=True)
dv.add_formula('DDIF2','-(DIF2-MACD2)',is_quarterly=False,add_data=True)
```

```python
#10、TVMA201    
dv.add_formula('TVMA201','-Ts_Mean(turnover,20)',is_quarterly=False,add_data=True)
```

```python
#11、PPEHIST601
dv.add_formula('PEHIST601','-PE/Ts_Mean(PE,60)', is_quarterly=False,add_data=True) 
```

```python
#12、PEEH1201 
dv.add_formula('PEEH1201','-PE/Ewma(PE,30)', is_quarterly=False,add_data=True) 
```

```python
#13、BIAS2_1
dv.add_formula('BIAS2_1','-(close-Ewma(close,12))/Ewma(close,12)*100',is_quarterly=False,add_data=True)
```
